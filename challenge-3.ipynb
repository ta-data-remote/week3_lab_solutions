## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

```python
# your code here
sum_amount = orders.pivot_table(index = ['CustomerID'], values = ['amount_spent', 'Country'], aggfunc = {'amount_spent':'sum', 'Country':'max'}) 
sum_amount.columns = ['Country', 'sum_amount_spent']
sum_amount
```

```python
bin_labels = ['Others', 'Preferred Customers','VIP Customers']

sum_amount['Customer Type'] = pd.qcut(sum_amount['sum_amount_spent'],
                q=[0, .75, .95, 1], 
                labels = bin_labels)

sum_amount.head()
```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here
qsum = sum_amount.pivot_table(index = ['Customer Type', 'Country'], values = ['Customer Type'], aggfunc = {'count'}).reset_index()
VIP = qsum[qsum['Customer Type'] == 'VIP Customers']
VIP.columns = ['Status', 'Country', 'VIP_count']
display(VIP)
VIP_max = VIP[VIP['VIP_count'] == VIP['VIP_count'].max()]
VIP_max
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
VIP_Pref = qsum[(qsum['Customer Type'] == 'VIP Customers') | (qsum['Customer Type'] == 'Preferred Customers')]
VIP_Pref.columns = ['Status', 'Country', 'status_count']
VIP_Pref = VIP_Pref.pivot_table(index = 'Country', values = ['status_count'], aggfunc = {'status_count':'sum'}) 
display(VIP_Pref)
VIP_Pref_max = VIP_Pref[VIP_Pref['status_count'] == VIP_Pref['status_count'].max()]
VIP_Pref_max
```
